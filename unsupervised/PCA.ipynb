```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
X, y = make_blobs(n_samples=500,n_features=5,centers=3,cluster_std=1.5,random_state=42)
```

```python
print(pd.DataFrame(X).value_counts())
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
pca = PCA(n_components=2)
Xpca = pca.fit_transform(X_scaled)
```

```python
Xpca
```

```python
pcadf = pd.DataFrame(Xpca,columns=["PCA1","PCA2"])
```

```python
pcadf
```

```python
pcadf['label'] = y
```

```python
sns.scatterplot(data=pcadf,x='PCA1',y="PCA2",hue=pcadf['label'],palette='Set2')
```
